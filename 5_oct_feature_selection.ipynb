```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris
```

```python
x = iris.data
x
```

```python
y = iris.target
y
```

```python
x
```

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
sel = VarianceThreshold(threshold=0.5)
```

```python
x_sel = sel.fit_transform(x)
```

```python
x_sel
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure(figsize=(8,10))
ax = fig.add_subplot(111,projection='3d')
ax.scatter3D(x_sel[:,0],x_sel[:,1],x_sel[:,2],c=y,cmap='rainbow')
plt.show()
```

```python
sel_2 = VarianceThreshold(threshold=0.6)
```

```python
x_sel2 = sel_2.fit_transform(x)
```

```python
x_sel2
```

```python
plt.scatter(x_sel2[:,0],x_sel2[:,1],c=y,cmap='rainbow')
plt.show()
```

```python
from sklearn.svm import LinearSVC
```

```python
model = LinearSVC()
```

```python
model.fit(x,y)
```

```python
from sklearn.feature_selection import SelectFromModel
```

```python
sel_model = SelectFromModel(model,prefit=True)
```

```python
x_sfm = sel_model.transform(x)
```

```python
x_sfm
```

```python
fig = plt.figure(figsize=(8,10))
ax = fig.add_subplot(111,projection='3d')
ax.scatter3D(x_sfm[:,0],x_sfm[:,1],x_sfm[:,2],c=y,cmap='rainbow')
plt.show()
```
