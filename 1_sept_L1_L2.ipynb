```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.array([-1,0,1,2,3,4])
```

```python
y = np.array([-3,-1,1,3,5,7])
```

```python
plt.scatter(x,y)
plt.xlabel('Independent')
plt.ylabel('Dependent')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
x
```

```python
x = x.reshape(-1,1)
```

```python
x.shape
```

```python
x
```

```python
model = LinearRegression()
```

```python
model.fit(x,y)
```

```python
model.score(x,y)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
plt.scatter(x,y)
plt.plot(x,model.predict(x),color='red')
plt.xlabel('Independent')
plt.ylabel('Dependent')
plt.show()
```

```python
# Lasso
```

```python
from sklearn.linear_model import Lasso
```

```python
model1 = Lasso(alpha=0.8)
```

```python
model1.fit(x,y)
```

```python
model1.score(x,y)
```

```python
plt.scatter(x,y)
plt.plot(x,model1.predict(x),color='red')
plt.xlabel('Independent')
plt.ylabel('Dependent')
plt.show()
```

```python
# Ridge
```
