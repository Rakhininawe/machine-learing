```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
salary = np.array([200,30,450,150,300])
level = np.array([1,2,3,4,5])
```

```python
fig = plt.figure(figsize=(8,10))
ax = fig.add_axes([0,0,1,1])
ax.plot(level,salary)
plt.show()
```

```python
incent = np.array([20,30,40,50,10])
```

```python
incent
```

```python
fig = plt.figure(figsize=(6,4))
ax = fig.add_axes([0,0,1,1])
ax.plot(level,salary,label="level vs salary")
ax.plot(level,incent,label="level vs incentives")
plt.legend(loc=0)
plt.show()
```
